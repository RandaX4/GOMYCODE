```python
import pandas as pd
import numpy as np

exam_date={'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas', "Elias", 'Myriam'],

'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan , 8, 19, 20, 16],

'attempts': [1, 3, 2, 3, 2, 3, 1, 1, np.nan ,1, 3, 2],

'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no',np.nan , 'yes', 'yes', 'yes']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l']
df=pd.DataFrame(exam_date,index=labels)
print(df)
```

```python
df.isnull().sum()
```

```python
df["score"].fillna(df["score"].mean(), inplace=True)
df
```

```python
df["attempts"].fillna(df["attempts"].mode(), inplace=True)
df
```

```python
df["qualify"]=df["qualify"].fillna("no")
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
df["name"]=encoder.fit_transform(df["name"])
df
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
df["qualify"]=encoder.fit_transform(df["qualify"])
df
```

```python
one_hot=pd.get_dummies(df["qualify"])
df.drop("qualify",axis=1,inplace=True)
df=df.join(one_hot)
df
```
